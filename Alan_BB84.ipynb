```python
import random

# Function to generate random bit string
def generate_bit_string(length):
    return [random.randint(0, 1) for _ in range(length)]

# Function to generate random basis string
def generate_basis_string(length):
    return [random.randint(0, 1) for _ in range(length)]

# Function to encode bit string based on basis
def encode_bits(bits, basis):
    encoded_bits = []
    for bit, b in zip(bits, basis):
        if b == 0:  # Rectilinear basis
            encoded_bits.append(bit)
        else:       # Diagonal basis
            encoded_bits.append((bit + 1) % 2)  # Apply NOT gate
    return encoded_bits

# Function to decode received bits based on basis
def decode_bits(bits, basis):
    decoded_bits = []
    for bit, b in zip(bits, basis):
        if b == 0:  # Rectilinear basis
            decoded_bits.append(bit)
        else:       # Diagonal basis
            decoded_bits.append((bit + 1) % 2)  # Apply NOT gate
    return decoded_bits

# Function to check for errors and establish shared key
def check_errors_and_generate_key(sent_bits, received_bits, basis):
    key = []
    for sent_bit, received_bit, b in zip(sent_bits, received_bits, basis):
        if b == 0:  # Rectilinear basis
            if sent_bit == received_bit:
                key.append(sent_bit)
        else:       # Diagonal basis
            if sent_bit == received_bit:
                key.append((sent_bit + 1) % 2)  # Apply NOT gate
    return key

# Alice's steps
def alice():
    n = 10  # Length of bit string

    # Alice generates random bit string and basis string
    alice_bits = generate_bit_string(n)
    alice_basis = generate_basis_string(n)
    
    # Alice encodes her bits based on her basis
    encoded_bits = encode_bits(alice_bits, alice_basis)
    
    return encoded_bits, alice_basis

# Bob's steps
def bob(encoded_bits, bob_basis):
    n = len(encoded_bits)
    
    # Bob generates his own basis string
    bob_basis = generate_basis_string(n)
    
    # Bob measures the encoded bits based on his basis
    bob_bits = decode_bits(encoded_bits, bob_basis)
    
    return bob_bits, bob_basis

# Communication between Alice and Bob
def main():
    # Alice sends encoded bits and basis to Bob
    encoded_bits, alice_basis = alice()
    
    # Bob receives encoded bits and basis
    bob_bits, bob_basis = bob(encoded_bits, alice_basis)
    
    # Alice and Bob compare basis and generate the shared key
    shared_key = check_errors_and_generate_key(encoded_bits, bob_bits, alice_basis)
    
    print("Shared Key:", shared_key)

if __name__ == "__main__":
    main()
```
